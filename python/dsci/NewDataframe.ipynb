```python
train = pd.read_csv('http://bit.ly/kaggletrain')
train.head()
```

```python
train['Sex_num'] = train.Sex.map({'female':0, 'male':1})
```

```python
train.loc[0:4, ['Sex', 'Sex_num']]
```

```python
train['Name_length'] = train.Name.apply(len)
```

```python
train.loc[0:4, ['Name', 'Name_length']]
```

```python
train.head()
```

```python
train['Fare_ceil'] = train.Fare.apply(np.ceil)
```

```python
train.loc[0:4, ['Fare', 'Fare_ceil']]
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
def getWorst(numDrinks):
    return np.argmax(numDrinks).split('_')[0].title()

drinks['Worst'] = drinks.loc[:, 'beer_servings':'wine_servings'].apply(getWorst, axis=1)
```

```python
drinks.loc[:, ['country', 'Worst']]
```

```python
drinks['Worst'] = drinks.loc[:, 'Worst'].apply(lambda s: s.split('_')[0].title())
```
