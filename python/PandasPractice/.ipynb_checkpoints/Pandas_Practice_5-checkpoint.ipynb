## Using iloc() and ix()

```python
import pandas as pd
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
ufo.head()
```

```python
ufo.iloc[:, [0, 3]]
```

```python
ufo[['City', 'State']]
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col='country')
drinks.head()
```

```python
drinks.ix['Albania', 0]
```

```python
drinks.ix[1, 'beer_servings']
```

```python
drinks.ix['Albania':'Andorra', 0:2]
```

```python
ufo.ix[0:2, 0:2]
```
